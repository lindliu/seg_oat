```python
import numpy as np
from cellpose import models, core, io, plot
from pathlib import Path
from tqdm import trange
import matplotlib.pyplot as plt

io.logger_setup() # run this to get printing of progress

#Check if colab notebook instance has GPU access
if core.use_gpu()==False:
  raise ImportError("No GPU access, change your runtime")

model = models.CellposeModel(gpu=True)
```

# dataset and train custom model

```python
train_dir = './data/annotation/train/'
test_dir = './data/annotation/test/'

masks_ext = "_seg.npy"
```

```python
from cellpose import train
import os.path

model_name = "new_model"

# default training params
n_epochs = 500
learning_rate = 1e-5
weight_decay = 0.1
batch_size = 1

# get files
output = io.load_train_test_data(train_dir, test_dir, mask_filter=masks_ext)
train_data, train_labels, _, test_data, test_labels, _ = output
# (not passing test data into function to speed up training)

if not os.path.isfile(f'./models/{model_name}'):
    new_model_path, train_losses, test_losses = train.train_seg(model.net,
                                                                train_data=train_data,
                                                                train_labels=train_labels,
                                                                batch_size=batch_size,
                                                                n_epochs=n_epochs,
                                                                learning_rate=learning_rate,
                                                                weight_decay=weight_decay,
                                                                nimg_per_epoch=max(2, len(train_data)), # can change this
                                                                model_name=model_name)
```

# validation

```python
from cellpose import metrics

new_model_path = './models/new_model'
model = models.CellposeModel(gpu=True,
                             pretrained_model=new_model_path)

# run model on test images
masks = model.eval(test_data, batch_size=1)[0]

# check performance using ground truth labels
ap = metrics.average_precision(test_labels, masks)[0]
print('')
print(f'>>> average precision at iou threshold 0.5 = {ap[:,0].mean():.3f}')
```

```python
import tifffile
import pickle

tifffile.imwrite('output.tiff', volume.astype(np.float32))

img_3D = io.imread("output.tiff")
```

```python
# 1. computes flows from 2D slices and combines into 3D flows to create masks
masks, flows, _ = model.eval(img_3D, z_axis=0, channel_axis=None,
                                batch_size=32,
                                do_3D=True, flow3D_smooth=1)

np.save('masks.npy', masks)
with open('flows.pkl', 'wb') as f:
    pickle.dump(flows, f)
```

```python
# 2. computes masks in 2D slices and stitches masks in 3D based on mask overlap
print('running cellpose 2D + stitching masks')
masks_stitched, flows_stitched, _ = model.eval(img_3D, z_axis=0, channel_axis=None,
                                                  batch_size=32,
                                                  do_3D=False, stitch_threshold=0.5)

np.save('masks_stitched.npy', masks_stitched)
with open('flows_stitched.pkl', 'wb') as f:
    pickle.dump(flows_stitched, f)

# with open('data.pkl', 'rb') as f:
#     loaded_data = pickle.load(f)
# print(loaded_data)
```

```python
import numpy as np
import copy 

# Get coordinates where value != 0
points = np.argwhere(masks > 0)  ## masks (Z,X,Y), points (Z,X,Y)
color = masks[points[:,0], points[:,1], points[:,2]]/np.max(masks)
```

```python
import open3d as o3d


# 自定义颜色，每个点对应一个 RGB 值（范围 [0, 1]）
colors = np.zeros_like(points, dtype=float)      # 所有点初始为黑色
colors[:, 0] = color          # 红色通道 = x 坐标
colors[:, 1] = color/1.5          # 绿色通道 = y 坐标
colors[:, 2] = color/2          # 蓝色通道 = z 坐标

# 构建点云
pcd = o3d.geometry.PointCloud()
pcd.points = o3d.utility.Vector3dVector(points)
pcd.colors = o3d.utility.Vector3dVector(colors)

# 可视化
o3d.visualization.draw_geometries([pcd])
```
